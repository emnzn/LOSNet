```python
alive=pd.read_csv('static_cleaned_dead.csv')
alive
```

```python
alive['multi_label']=alive['los_icu'].apply(lambda x:0 if x<2 else(1 if 2<=x<=4 else 2))
alive.drop(['hosp_admittime','hosp_dischtime','icu_intime','icu_outtime','los_icu_binned'],axis=1,inplace=True)
alive.drop(['los_icu'],axis=1,inplace=True)
alive=pd.get_dummies(alive)
alive
```

```python
alive.to_csv('st_dead_pcs.csv')
```

```python
X=alive.drop(['multi_label'],axis=1)
X
```

```python
y=alive['multi_label']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
alive_train=pd.concat([X_train,y_train],axis=1)
alive_train.to_csv('st_dead_train.csv')
```

```python
alive_test=pd.concat([X_test,y_test],axis=1)
alive_test.to_csv('st_dead_test.csv')
```

```python
dy=pd.read_csv('dynamic_cleaned.csv')
dy_train = dy[dy['id'].isin(alive_train['id'])]
dy_train.to_csv('dynamic_dead_train.csv')
```

```python
dy_test = dy[dy['id'].isin(alive_test['id'])]
dy_test.to_csv('dynamic_dead_test.csv')
```
