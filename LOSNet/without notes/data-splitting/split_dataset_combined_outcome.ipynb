```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
notes = pd.read_csv('../../data/notes_cleaned.csv')

notes.head()
```

```python
dynamic = pd.read_csv('../../data/dynamic_cleaned.csv')

dynamic.head()
```

```python
static = pd.read_csv('../../data/static_cleaned.csv')
static = static[static['id'].isin(dynamic['id'])]
static.head()
```

```python
static.shape
```

```python
base_split_path = '../../data/split/with-outliers/combined'
```

```python
static_train, static_val = train_test_split(
    static, test_size=0.20, 
    random_state=42, stratify=static['los_icu_binned']
    )


static_train.to_csv(f'{base_split_path}/static_train.csv', index=False)
static_val.to_csv(f'{base_split_path}/static_val.csv', index=False)

print('saved')
```
